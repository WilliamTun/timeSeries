# Scipy package

```python
from scipy.signal import find_peaks
```

```python
def plotIt(x, y, peak_index):
    plt.figure(figsize=(20,10))
    plt.subplot(2, 2, 1)
    plt.plot(peak_index, y[peak_index], "xr", markersize=20); 
    plt.plot(y); plt.legend(['distance'])
```

```python
peaks_index, _ = find_peaks(y, distance=20)
plotIt(x, y, peaks_index)
```

```python
peaks_index2, _ = find_peaks(y, prominence=1)
plotIt(x, y, peaks_index2)
```

```python
peaks_index3, _ = find_peaks(y, width=20)
plotIt(x, y, peaks_index3)
```

```python
peaks_index4, _ = find_peaks(y, threshold=0.4)
plotIt(x, y, peaks_index4)
```

# Peak Detect

```python
from peakdetect import peakdetect
```

```python
peaks = peakdetect(y, lookahead=20) 
```

```python
peaks
```

```python
higherPeaks = np.array(peaks[0])
lowerPeaks = np.array(peaks[1])
```

```python
higherPeakIndexes = []
for i in higherPeaks:
    higherPeakIndexes.append(i[0])
```

```python
lowerPeakIndexes = []
for i in lowerPeaks:
    lowerPeakIndexes.append(i[0])
```

```python
plotIt(x, y, lowerPeakIndexes)
```

```python
plotIt(x, y, higherPeakIndexes)
```
